```python
#GrupoB Interconsultas
def conteo(grupo,piso):
    #Crear el grupo B filtrado por pisos
    df_MI_grupoB=piso
    df_MI_grupoB.reset_index(drop=True)
    #Borrramos las notas adicionales 
    #indexNames = df_MI_grupoB[ df_MI['Tipone'] =='notas'].index
   # df_MI_grupoB.drop(indexNames , inplace=True)
    pac_grupoB=df_MI_grupoB['Fecha'].value_counts(dropna=True)
    #le damos formato a las variables
    pac_grupoB = pd.DataFrame(pac_grupoB)
    pac_grupoB = pac_grupoB.reset_index()
    pac_grupoB.columns = ['Fecha', 'numero_pacientes']
    pac_grupoB.set_index('Fecha')

    #cuenta los pacientes por medico en el grupo  y los agrega al DF

    Medico=df_MI_grupoB['Medico']==grupo[0]
    df_Medico=df_MI_grupoB[Medico]
    df_Medico.set_index('Fecha')
    df_count=df_Medico.groupby('Fecha').count()
    Medicodf=df_count['NotaEvolucionId']
    Medicodf=pd.DataFrame(Medicodf)
    Medicodf.columns = [grupo[0]]
    Grupo_B=pd.merge(Medicodf,pac_grupoB, indicator=False, how='right', on='Fecha')

    #cuenta los pacientes por medico en el grupo  y los agrega al DF

    Medico=df_MI_grupoB['Medico']==grupo[1]
    df_Medico=df_MI_grupoB[Medico]
    df_Medico.set_index('Fecha')
    df_count=df_Medico.groupby('Fecha').count()
    Medicodf=df_count['NotaEvolucionId']
    Medicodf=pd.DataFrame(Medicodf)
    Medicodf.columns = [grupo[1]]
    Grupo_B1=pd.merge(Medicodf,Grupo_B, indicator=False, how='right', on='Fecha')
    #cuenta los pacientes por medico en el grupo  y los agrega al DF

    Medico=df_MI_grupoB['Medico']==grupo[2]
    df_Medico=df_MI_grupoB[Medico]
    df_Medico.set_index('Fecha')
    df_count=df_Medico.groupby('Fecha').count()
    Medicodf=df_count['NotaEvolucionId']
    Medicodf=pd.DataFrame(Medicodf)
    Medicodf.columns = [grupo[2]]
    Grupo_B2=pd.merge(Medicodf,Grupo_B1, indicator=False, how='right', on='Fecha')
    #cuenta los pacientes por medico en el grupo  y los agrega al DF

    Medico=df_MI_grupoB['Medico']==grupo[3]
    df_Medico=df_MI_grupoB[Medico]
    df_Medico.set_index('Fecha')
    df_count=df_Medico.groupby('Fecha').count()
    Medicodf=df_count['NotaEvolucionId']
    Medicodf=pd.DataFrame(Medicodf)
    Medicodf.columns = [grupo[3]]
    Grupo_B3=pd.merge(Medicodf,Grupo_B2, indicator=False, how='right', on='Fecha')
    #cuenta los pacientes por medico en el grupo  y los agrega al DF

    Medico=df_MI_grupoB['Medico']==grupo[4]
    df_Medico=df_MI_grupoB[Medico]
    df_Medico.set_index('Fecha')
    df_count=df_Medico.groupby('Fecha').count()
    Medicodf=df_count['NotaEvolucionId']
    Medicodf=pd.DataFrame(Medicodf)
    Medicodf.columns = [grupo[4]]
    Grupo_B4=pd.merge(Medicodf,Grupo_B3, indicator=False, how='right', on='Fecha')

    #cuenta los pacientes por medico en el grupo  y los agrega al DF

    Medico=df_MI_grupoB['Medico']==grupo[5]
    df_Medico=df_MI_grupoB[Medico]
    df_Medico.set_index('Fecha')
    df_count=df_Medico.groupby('Fecha').count()
    Medicodf=df_count['NotaEvolucionId']
    Medicodf=pd.DataFrame(Medicodf)
    Medicodf.columns = [grupo[5]]
    Grupo_B5=pd.merge(Medicodf,Grupo_B4, indicator=False, how='right', on='Fecha')
    Grupo_IC=Grupo_B5.set_index('Fecha')
    
    try:
        Medico=df_MI_grupoB['Medico']==grupo[6]
        df_Medico=df_MI_grupoB[Medico]
        df_Medico.set_index('Fecha')
        df_count=df_Medico.groupby('Fecha').count()
        Medicodf=df_count['NotaEvolucionId']
        Medicodf=pd.DataFrame(Medicodf)
        Medicodf.columns = [grupo[6]]
        Grupo_B6=pd.merge(Medicodf,Grupo_B5, indicator=False, how='right', on='Fecha')
        Grupo_IC=Grupo_B6.set_index('Fecha')
    except:
        pass
    try:
        Medico=df_MI_grupoB['Medico']==grupo[7]
        df_Medico=df_MI_grupoB[Medico]
        df_Medico.set_index('Fecha')
        df_count=df_Medico.groupby('Fecha').count()
        Medicodf=df_count['NotaEvolucionId']
        Medicodf=pd.DataFrame(Medicodf)
        Medicodf.columns = [grupo[7]]
        Grupo_B7=pd.merge(Medicodf,Grupo_B6, indicator=False, how='right', on='Fecha')
        Grupo_IC=Grupo_B7.set_index('Fecha')
    except:
        pass
    return Grupo_IC

```
